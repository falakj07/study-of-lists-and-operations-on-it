```python
## Access Items
mylist = ["jahnvi","hiranya","falak"]
print(mylist[0])
```

```python
## Revrese/Negative Indexing
mylist = ["jahnvi","hiranya","falak"]
print(mylist[-3])
```

```python
## Slicing
mylist = ["jahnvi","hiranya","falak" , "rivya", "divisha"]
print(mylist[-3:])
```

```python
mylist = ["jahnvi","hiranya","falak" , "rivya", "divisha"]
print(mylist[-3:-1])
```

```python
mylist = ["jahnvi","hiranya","falak" , "rivya", "divisha"]
print(mylist[-3:])
```

```python
## Change Item Value
mylist = ["jahnvi","hiranya","falak" , "rivya", "divisha"]
mylist[1]= "darsh"
print(mylist)
```

```python
## Append items
mylist = ["jahnvi","hiranya","falak" , "rivya", "divisha"]
mylist.append("darsh")
print(mylist)
```

```python
## Insert items
mylist = ["jahnvi","hiranya","falak" , "rivya", "divisha"]
mylist.insert(3,"aadi")
print(mylist)
```

```python
## Extend list
mylist = ["jahnvi","hiranya","falak" , "rivya", "divisha"]
people = ["darsh ","aadi","viraj"]
mylist.extend (people)
print(mylist)
```
